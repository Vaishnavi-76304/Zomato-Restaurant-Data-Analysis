```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Data Cleaning
df = pd.read_csv('/content/zomato.csv')
df
```

```python
df=df.rename(columns={'listed_in(type)':'Type','listed_in(city)':'city'})
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.dropna()
df
```

```python
df.rename(columns={'url': 'URL',
                        'address': 'ADDRESS',
                        'name' : 'NAME',
                        'online_order':'ONLINE_ORDER',
                        'book_table': 'BOOK_TABLE',
                        'rate': 'RATE', 'votes': 'VOTES',
                        'phone': 'PHONE','location': 'LOCATION','rest_type': 'REST_TYPE',
                        'dish_liked': 'DISH_LIKED',
                        'cuisines': 'CUISINES',
                        'approx_cost(for two people)':'APPROX_COST',
                        'reviews_list': 'REVIEW_LIST','menu_item': 'MENU_ITEM',
                        'Type': 'TYPE','city': 'CITY'},inplace= True)
df.head()
```

```python
df['RATE']= df['RATE'].str.split('/5').str[0]
df
```

```python
df['RATE'].unique()
```

```python
df['RATE'] = df['RATE'].replace('NEW',0)
df
```

```python
df['RATE'].unique()
```

```python
df['RATE']= df['RATE'].astype(float)
```

```python
df['APPROX_COST'] = df['APPROX_COST'].str.replace(',','')
```

```python
df['APPROX_COST']=df['APPROX_COST'].astype(float)
```

```python
#Data Visualzaion
# Restaurants delivering Online or not
df['ONLINE_ORDER'].value_counts().plot(kind= 'bar')
```

```python
# Restaurants allowing table booking or not
df['BOOK_TABLE'].value_counts().plot(kind= 'bar')
```

```python
# Table booking Rate vs Rate
import seaborn as sns
sns.scatterplot(x= 'APPROX_COST',y= 'RATE', data= df)
```

```python
# Best Location
df['LOCATION'].value_counts().plot(kind= 'bar')
```

```python
# Relation between Location and Rating
sns.boxplot(x= 'LOCATION',y= 'RATE',data= df)
```

```python
# Restaurant Type
df['REST_TYPE'].value_counts().plot(kind= 'bar')
```

```python
# Gaussian Rest type and Rating
sns.kdeplot(x= 'RATE',hue= 'REST_TYPE', data= df)
```

```python
# Types of Services
df['TYPE'].value_counts().plot(kind= 'bar')
```

```python
# Relation between Type and Rating
sns.boxplot(x='TYPE',y= 'RATE', data= df)
```
